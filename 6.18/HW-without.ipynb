# PCA 주성분 분석

#### 각 변수에서 주된 성분을 뽑아서 분석해준다.

#### pca는 자체적으로 공분산값을 구해서 아이겐벨류값을 얻어서 그걸 비교해서 위에서 부터 큰값을 리턴한다


#### pca는 값을 타겟하는 과정이  필요없는 함수이다. 따라서 위의 과정처럼 `y=df['total_bill'] ` 과 `X=df.drop('total_bill',axis=1)`가 필요없다. 그래서 `y=df['total_bill'] ` 과 `X=df.drop('total_bill',axis=1)`의 영향이 없는 `datatranss`를 이용하였다.

```python
from sklearn.decomposition import PCA
```

```python
# define the transform
trans = PCA(n_components=4)  #주성분을 4개 골라준다.
```

```python
# transform the data
X_dim = trans.fit_transform(datatranss)
```

```python
# summarize data after the transform
print(X_dim[:3, :]) #X_dim은 array라서 describe(),info() 불가
```

```python
trans.components_  
```

```python
trans.explained_variance_           #주성분의 분산값
```

```python
trans.explained_variance_ratio_     #주성분의 분산값의 비율
```

```python
trans.explained_variance_ratio_.cumsum() #각 행마다의 데이터 분포값을 더해주기
```

```python
trans.explained_variance_ratio_.sum()     #주성분의 분산값의 비율의 총합
                                          #전체 데이터에서 얼마를 대표하는 지 
```

```python
x_dim=pd.DataFrame(X_dim)
```

```python
x_dim.describe().round()
```

```python
x_dim.profile_report()
```

# Regression Feature Selection

#### 수치형 종속변수와 수치형 독립변수사이의 영향 또는 인과관계를 알 수 있는 분석
#### 상관관계 -> 인과관계로 바꿔주는 함수.(통계학과 수업시간에 배움..)

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_regression
```

```python
# define feature selection
fs = SelectKBest(score_func=f_regression, k=4)  #7개 변수 중 4개만 취한다.
```

```python
datatrans.shape
```

```python
y.shape
```

```python
# apply feature selection / 행이 같으므로 값이 나온다.
X_selected = fs.fit_transform(datatrans, y)
```

```python
fs.get_support(indices=True)   # 6개의 행 중에서 0,1,4,5만 취한다.
```

```python
fs.get_support(indices=False)  # 6개의 행 중에서 2,3을 뺀다.
```
