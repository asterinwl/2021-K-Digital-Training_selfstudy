```python
fig, axs = plt.subplots(nrows=2,ncols= 2)   #이 경우는 fig를 안써서 _,axs로 두어도 괜찮다.
axs[0,0].plot([1, 2, 3, 4], [1, 3, 4, 2])
axs[1,1].plot([1, 2, 3, 4], [1, 4, 2, 3])
```

```python
_, axs = plt.subplots(nrows=2,ncols= 2)   #이 경우는 fig를 안써서 _,axs로 두어도 괜찮다.
axs[0,0].plot([1, 2, 3, 4], [1, 3, 4, 2])
axs[1,1].plot([1, 2, 3, 4], [1, 4, 2, 3])
```

```python
fig, axs = plt.subplots(nrows=2,ncols= 2)   
axs[0,0].plot([1, 2, 3, 4], [1, 3, 4, 2])
axs[1,1].plot([1, 2, 3, 4], [1, 4, 2, 3])
fig.suptitle('Vertically stacked subplots')
```

```python
fig, axs = plt.subplots(nrows=2,ncols= 2)   
axs[0,0].plot([1, 2, 3, 4], [1, 3, 4, 2])
axs[1,1].plot([1, 2, 3, 4], [1, 4, 2, 3])
fig.suptitle('Vertically stacked subplots')
fig.supxlabel('X label')
fig.supylabel('Y label')
axs[1,1].set_title('Simple plot')
```

```python
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(a, a, label='linear')  # Plot some data on the axes.
ax.plot(a, a**2, label='quadratic')  # Plot more data on the axes...
ax.plot(a, a**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend. ()안에 shift+Tab을 누른 후 설명 읽고, 위치/폰트크기/색상 등 변경
```
